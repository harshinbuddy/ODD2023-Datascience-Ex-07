```python
import pandas as pd

# Assuming your dataset is a CSV file, replace 'your_dataset.csv' with the actual file path or URL
file_path = 'bank_marketing.csv'
df = pd.read_csv(file_path)

# Filter the DataFrame to include only clients who have subscribed to a deposit
deposit_subscribers = df[df['deposit'] == 'yes']

# Filter the resulting DataFrame to include only clients who have either a housing or personal loan
clients_with_loan = deposit_subscribers[(deposit_subscribers['housing'] == 'yes') | (deposit_subscribers['loan'] == 'yes')]

# Display the number of clients who have subscribed to a deposit and have either a housing or personal loan
num_clients_with_loan = len(clients_with_loan)
print(f"Number of clients who have subscribed to a deposit and have either a housing or personal loan: {num_clients_with_loan}")
```

```python
import pandas as pd

# Assuming your dataset is a CSV file, replace 'your_dataset.csv' with the actual file path or URL
file_path = 'bank_marketing.csv'
df = pd.read_csv(file_path)

# Filter the DataFrame to include only clients who have subscribed to a deposit
deposit_subscribers = df[df['deposit'] == 'yes']

# Calculate the average age of clients who have subscribed to a deposit
avg_age_deposit_subscribers = deposit_subscribers['age'].mean()

print(f"Average age of clients who have subscribed to a deposit: {avg_age_deposit_subscribers:.2f} years")
```

```python
import pandas as pd

# Assuming your dataset is a CSV file, replace 'your_dataset.csv' with the actual file path or URL
file_path = 'bank_marketing.csv'
df = pd.read_csv(file_path)

# Filter the DataFrame to include only clients who have not subscribed to a deposit
not_subscribed_clients = df[df['deposit'] == 'no']

# Find the maximum number of contacts for clients who have not subscribed to a deposit
max_contacts_not_subscribed = not_subscribed_clients['campaign'].max()

print(f"Maximum number of contacts for clients who have not subscribed to a deposit: {max_contacts_not_subscribed}")
```

```python
import pandas as pd

# Assuming your dataset is a CSV file, replace 'your_dataset.csv' with the actual file path or URL
file_path = 'bank_marketing.csv'
df = pd.read_csv(file_path)

# Find the maximum balance for clients who have subscribed to a deposit
max_balance_subscribed = df[df['deposit'] == 'yes']['balance'].max()

# Find the maximum balance for clients who have not subscribed to a deposit
max_balance_not_subscribed = df[df['deposit'] == 'no']['balance'].max()

# Calculate the difference in maximum balance
balance_difference = max_balance_subscribed - max_balance_not_subscribed

print(f"Difference between the maximum balance for subscribed and not subscribed clients: {balance_difference:.2f} euros")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataset is a CSV file, replace 'your_dataset.csv' with the actual file path or URL
file_path = 'bank_marketing.csv'
df = pd.read_csv(file_path)

# Generate a scatter plot of "age" vs "balance"
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['balance'], alpha=0.5)
plt.title('Scatter Plot of Age vs Balance')
plt.xlabel('Age')
plt.ylabel('Balance (euros)')
plt.show()
```
